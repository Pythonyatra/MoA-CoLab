```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

import random
import time
```

```python
train = pd.read_csv('./datasets/train_features.csv')
test = pd.read_csv('./datasets/test_features.csv')
```

```python
print(train.shape)
print(test.shape)
```

```python
train.head()
```

```python
train.describe()
```

```python
test.head()
```

```python
test.describe()
```

```python
train['dataset'] = 'train'
test['dataset']='test'
```

```python
print(train.head(2))
print(test.head(2))
```

```python
data = pd.concat([train, test])
data.head()
```

**cp_type feature** 

```python
ctype = data.groupby(['cp_type','dataset'])['sig_id'].count().reset_index()
ctype.columns=['cp_type', 'dataset', 'count']
ctype.head()
```

```python
fig = px.bar(
    ctype, 
    x='cp_type', 
    y="count", 
    color = 'dataset',
    barmode='stack',
    orientation='v', 
    title='Feature cp_type counts', 
    width=500,  
    height=400  
)
fig.show(renderer='svg')
```

**cp_time feature**

```python
cptime = data.groupby(['cp_time', 'dataset'])['sig_id'].count().reset_index()
cptime.columns = ['cp_time', 'dataset', 'count']
cptime.head()
```

```python
fig = px.bar(
    cptime, 
    x='cp_time', 
    y="count", 
    color = 'dataset',
    barmode='stack',
    orientation='v', 
    title='Feature cp_type Counts', 
    width=500,  
    height=400  
)
fig.show(renderer='svg')
```

**cp_dose feature**

```python
cpdose = data.groupby(['cp_dose', 'dataset'])['sig_id'].count().reset_index()
cpdose.columns = ['cp_dose', 'dataset', 'count']
cpdose.head()
```

```python
fig = px.bar(
    cpdose, 
    x='cp_dose', 
    y="count", 
    color = 'dataset',
    barmode='stack',
    orientation='v', 
    title='Feature cp_dose counts', 
    width=500,
    height=400
)

fig.show(renderer='svg')
```

```python
train_columns = train.columns.to_list()
gfeatures = [g for g in train_columns if g.startswith('g-')]
cfeatures = [c for c in train_columns if c.startswith('c-')]
restfeatures = [r for r in train_columns if (r[1]!= '-')]
```

```python
print(f"Total number of gene expression data features: {len(gfeatures)}")
print(f"Total number of cell viability data features: {len(cfeatures)}")
print(f"Rest of the features: {len(restfeatures)}" + f", and they are: {restfeatures}")
```
